```python
import numpy as np 
import pandas as pd 
```

```python
cust = pd.read_csv('olist_customers_dataset.csv')
geoloc = pd.read_csv('olist_geolocation_dataset.csv')
ord_items = pd.read_csv('olist_order_items_dataset.csv')
ord_payments = pd.read_csv('olist_order_payments_dataset.csv')
ord_reviews = pd.read_csv('olist_order_reviews_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')
sellers = pd.read_csv('olist_sellers_dataset.csv')
cat = pd.read_csv('product_category_name_translation.csv')
```

```python
list_df = [cust,
geoloc,
ord_items,
ord_payments,
ord_reviews,
orders,
products,
sellers,
cat]

Name = ['Customer','Geolocation','Order Items','Order Payments','Order Reviews','Orders','Products','Sellers','Category Translation']

for i,j in zip(list_df,Name):
    print('Table:',j)
    display(i.head(1),i.shape)
    print('\n\n')
```

# Merged Dataset : Outer Join

```python
merged_outer = pd.merge(orders, ord_reviews, on=['order_id'],how='outer')\
    .merge(ord_payments, on=['order_id'], how='outer')\
    .merge(ord_items, on=['order_id'],how='outer')\
    .merge(products, on=['product_id'],how='outer')\
    .merge(sellers, on=['seller_id'],how='outer')\
    .merge(cust, on=['customer_id'],how='outer')\
    .merge(cat, on=['product_category_name'],how='outer')

merged_outer
```

# Merged Dataset : Inner Joins

```python
merged_inner = pd.merge(orders, ord_reviews, on=['order_id'])\
    .merge(ord_payments, on=['order_id'])\
    .merge(ord_items, on=['order_id'])\
    .merge(products, on=['product_id'])\
    .merge(sellers, on=['seller_id'])\
    .merge(cust, on=['customer_id'])\
    .merge(cat, on=['product_category_name'])

merged_inner
```
